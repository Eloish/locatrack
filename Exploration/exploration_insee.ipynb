```python
# Regarder les valeurs de Q221 (revenu médian)
print(df["Q221"].value_counts().head(20))

# Combien de communes ont le secret statistique ?
print(f"\nCommunes avec secret statistique sur Q221 : {(df['Q221'] == 's').sum()}")
print(f"Communes avec revenu médian disponible : {(df['Q221'] != 's').sum()}")
```

```python
# Remplacer "s" par NaN et convertir en float
df["revenu_median"] = pd.to_numeric(
    df["Q221"].replace("s", None),
    errors="coerce"
)

# Vérification
print(df["revenu_median"].describe())
print(f"\nNulls : {df['revenu_median'].isna().sum()}")
```

```python
cols_utiles = [
    "CODGEO",        # code INSEE commune
    "Q221",          # revenu médian brut
    "D121",          # 1er décile
    "D921",          # 9ème décile
]

df_cols_utiles = df[cols_utiles].copy()

# Convertir toutes les colonnes numériques
for col in ["Q221", "D121", "D921"]:
    df_cols_utiles[col] = pd.to_numeric(
        df_cols_utiles[col].replace("s", None),
        errors="coerce"
    )

# Renommer pour plus de clarté
df_cols_utiles = df_cols_utiles.rename(columns={
    "CODGEO": "code_insee",
    "Q221": "revenu_median",
    "D121": "revenu_d1",
    "D921": "revenu_d9"
})

print(df_cols_utiles.dtypes)
print(df_cols_utiles.describe())
```

```python
# Vérification cohérence entre années
annees = [2017, 2018, 2019, 2020, 2021]

for annee in annees:
    df_tmp = pd.read_parquet(f"../data/bronze/insee/annee={annee}/insee_{annee}.parquet")
    a_q2 = "Q2" + str(annee)[-2:] in df_tmp.columns
    print(f"{annee} → {df_tmp.shape} | colonne revenu médian Q2{str(annee)[-2:]} : {a_q2}")
```
