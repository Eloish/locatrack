Comprendre les nulls sur loyer_mensuel_median

```python
# Regarder les lignes où loyer_mensuel_median est null
df_null = df[df["loyer_mensuel_median"].isna()]
print(df_null[["agglomeration", "Type_habitat", "nombre_pieces_homogene", "nombre_observations"]].head(20))
```

Filtrage

```python
# Je garde seulement les lignes qui ont un loyer médian
df_clean = df[df["loyer_mensuel_median"].notna()].copy()

print(f"Avant : {len(df)} lignes")
print(f"Après : {len(df_clean)} lignes")
```

```python
df_clean['Type_habitat'].isnull().
```

Vérifier le niveau géographique

Colonnes utiles

```python
cols_utiles = [
    "Observatory",
    "Data_year",
    "agglomeration",
    "Type_habitat",
    "nombre_pieces_homogene",
    "loyer_median",
    "loyer_mensuel_median",
    "surface_moyenne",
    "nombre_observations",
]

df_cols_utiles = df_clean[cols_utiles].copy()
print(df_cols_utiles.shape)
df_cols_utiles.head()
```

 Graphiques Plotly

```python
import plotly.express as px

# Distribution du loyer mensuel médian
fig = px.histogram(
    df_cols_utiles,
    x="loyer_mensuel_median",
    color="Type_habitat",
    nbins=50,
    title="Distribution du loyer mensuel médian par type de bien",
    labels={
        "loyer_mensuel_median": "Loyer mensuel médian (€)",
        "Type_habitat": "Type de bien"
    }
)
fig.show()

# Boxplot loyer mensuel par type de bien
fig = px.box(
    df_cols_utiles,
    x="Type_habitat",
    y="loyer_mensuel_median",
    color="Type_habitat",
    title="Distribution du loyer mensuel médian par type de bien",
    labels={
        "loyer_mensuel_median": "Loyer mensuel médian (€)",
        "Type_habitat": "Type de bien"
    }
)
fig.show()

# Top 10 agglomérations par loyer médian
top10 = (
    df_cols_utiles.groupby("agglomeration")["loyer_mensuel_median"]
    .median()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

fig = px.bar(
    top10,
    x="agglomeration",
    y="loyer_mensuel_median",
    title="Top 10 agglomérations par loyer mensuel médian",
    labels={
        "loyer_mensuel_median": "Loyer médian (€)",
        "agglomeration": "Agglomération"
    },
    color="loyer_mensuel_median",
    color_continuous_scale="Blues"
)
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
# Quelles agglomérations ont les loyers les plus élevés ?
df_top = df_clean[df_clean["loyer_mensuel_median"] > 957].sort_values(
    "loyer_mensuel_median", ascending=False
)
print(df_top[["agglomeration", "Type_habitat", "loyer_mensuel_median", "nombre_observations"]].head(20))
```
